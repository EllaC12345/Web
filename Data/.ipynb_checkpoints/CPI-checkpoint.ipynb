```python
## Create a New App 
def create_app():
    app = Flask(__name__)
    app.config['SECRET_KEY'] = "Simple Secret Key"
    ## Adding and configurring a SQLLite database for the Flask application
    app.config['SQLALCHEMY_DATABASE_URI'] = f'sqlite:///{Finance_db}'
    db.init_app(app) ## initialize the SQLAlchemy extension for Flask APP
    # Adding routes
    from .routes import routes
    app.register_blueprint(routes, url_prefix="/")
    ## Create Database
    create_database(app)
    return app
```
