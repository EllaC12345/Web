```python
import pandas as pd
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from yahoo_fin.stock_info import get_data
```

```python
## Import NYSE stock from CSV
NYSE_stock_df = pd.read_csv('/Users/ellandalla/Desktop/website Project/Data/NYSE_stocks.csv')
```

```python
NYSE_stock_df.head()
```

```python
stock_df_results = pd.DataFrame()
# Use a for loop to fetch historical data from YFinance
for symbol in NYSE_stock_df['Symbol']:
    try:
        data = get_data(symbol, start_date="2014-01-01", end_date="2023-07-31", index_as_date = True, interval="1mo")
        if not data.empty:
            stock_df_results = pd.concat([stock_df_results, data])
        else:
            print(f"No data available for {symbol}")
        # Process the data here
    except Exception as e:
        print(f"Error fetching data for {symbol}: {str(e)}")
        
      
```

```python
stock_df_results.iloc[:5]
```

```python
## Calculating 10 days Moving Average

stock_df_results['10 Day Moving Average'] = stock_df_results['close'].rolling(window=10).mean()
```

```python
## Determine the CAGR Rate 
# Calculate initial and final values
P0 = stock_df_results['close'].iloc[0]
Pn = stock_df_results['close'].iloc[-1]

## Determine the number of periods
n = len(stock_df_results)

# Calculate CAGR
CAGR = (Pn / P0) ** (1 / n) - 1

# Convert CAGR to a percentage
CAGR_percentage = CAGR * 100
stock_df_results['CAGR'] = CAGR_percentage
```
